### **Missing Value**

```python
from statistics import mode
```

```python
missing_percentage = (data.isnull().sum() / len(data)) * 100
columns_above_20_percent_missing = missing_percentage[missing_percentage >20]
columns_above_20_percent_missing
```

hapus kolom yang punya missing value >=20%

```python
X_train = X_train.drop(columns = ['Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'])
X_test = X_test.drop(columns = ['Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'])
```

```python
## handling missing value dengan persetase missing <20%
percent_missing = missing_percentage[missing_percentage> 0]
percent_missing.index
```

```python
numeric_features = X_train.select_dtypes(include=['int64','float64']).columns
categorical_features = X_train.select_dtypes(include=['object']).columns
```

```python
#handling missing value numerical column

col_missing_num = [col for col in numeric_features if col in percent_missing.index]

for i in col_missing_num:
    median = X_train[i].median()
    X_train[i] = X_train[i].fillna(median)
    X_test[i] = X_test[i].fillna(median)



#handling missing value categorical column
col_missing_cat = [col for col in categorical_features if col in percent_missing.index]

for i in col_missing_cat:
    mode_data = mode(list(X_train[i]))
    X_train[i] = X_train[i].fillna(mode_data)
    X_test[i] =X_test[i].fillna(mode_data)


```

```python
le = LabelEncoder()

for col in categorical_features:
    data[col]= le.fit_transform(data[col])
    X_train[col] = le.transform(X_train[col])
    X_test[col] = le.transform(X_test[col])


```

## **Modelling**

```python
import mlflow
import mlflow.sklearn
```

```python
mlflow.set_tracking_uri('sqlite:///mlflow.db')
mlflow.set_experiment('houseprice_prediction')

def evaluate_model(y_true, y_pred ): 
    """Calculate evaluation metrics"""
    return {
        'R2 Score': r2_score(y_true, y_pred),
        'MSE': mean_squared_error(y_true, y_pred),
        'RMSE': sqrt(mean_squared_error(y_true, y_pred)),
        'MAE': mean_absolute_error(y_true, y_pred),
        'MAPE': np.mean(np.abs((y_true - y_pred) / y_true))

    }

```

```python
models = {
    'decision_tree': {
        'model': DecisionTreeRegressor(),
        'params': {
            'max_depth': 10,  
            'min_samples_split': 2,
            'random_state': 42
        }
    },
    'random_forest': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': 100,
            'max_depth': 10,
            'random_state': 42
        }
    },
    'xgboost': {
        'model': XGBRegressor(),
        'params': {
            'n_estimators': 100,
            'learning_rate': 0.1,
            'max_depth': 3,
            'random_state': 42,
            'objective': 'reg:squarederror'  
        }
    }
}
```

```python
# Train and evaluate models
for model_name, model_info in models.items():
    with mlflow.start_run(run_name=model_name):
        # Log model parameters
        mlflow.log_params(model_info['params'])
        
        # Train model
        model = model_info['model']
        model.set_params(**model_info['params'])
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        # Calculate and log metrics
        metrics = evaluate_model(y_test, y_pred)
        mlflow.log_metrics(metrics)
        
        # Log model
        mlflow.sklearn.log_model(model, 
                                 model_name,
                                 registered_model_name=f"houseprice_{model_name}")
        
        print(f"\nModel: {model_name}")
        for metric_name, value in metrics.items():
            print(f"{metric_name}: {value:.4f}")
```

```python
# Transition the best model to production
from mlflow.tracking import MlflowClient

def load_production_model(model_name):
    model = mlflow.pyfunc.load_model(
        model_uri=f"models:/houseprice_{model_name}/Production"
    )
    return model



client = MlflowClient()

def transition_model_to_production(model_name):
    client = MlflowClient()
    latest_version = client.get_latest_versions(f"houseprice_{model_name}", stages=["None"])[0]
    client.transition_model_version_stage(
        name=f"houseprice_{model_name}",
        version=latest_version.version,
        stage="Production"
    )
```

```python
transition_model_to_production('xgboost')
```

```python
def get_all_runs():
    experiment = mlflow.get_experiment_by_name('houseprice_prediction')
    runs = mlflow.search_runs(experiment_ids=[experiment.experiment_id])
    return runs
```

```python
runs_df = get_all_runs()
metrics_comparison = runs_df[['tags.mlflow.runName', 'metrics.RMSE', 'metrics.MAE',
                              'metrics.R2 Score']]
print("\nModel Performance Comparison:")
print(metrics_comparison)
```
